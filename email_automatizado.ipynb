# Automação de Sistemas e Processos com Python

### Desafio:

Todos os dias, o nosso sistema atualiza as vendas do dia anterior.
O seu trabalho diário, como analista, é enviar um e-mail para a diretoria, assim que começar a trabalhar, com o faturamento e a quantidade de produtos vendidos no dia anterior

E-mail da diretoria: seugmail+diretoria@gmail.com<br>
Local onde o sistema disponibiliza as vendas do dia anterior: https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing

Para resolver isso, vamos usar o pyautogui, uma biblioteca de automação de comandos do mouse e do teclado

```python
# !pip install pyautogui
#openpyxl
#pandas
#numpy
```

```python
import pyautogui
import pyperclip
import time
```

```python
#Passo 1: Entrar no sistema da empresa (no exemplo é o link do drive = )

#Definir tempo de pausa [em segundos]
pyautogui.PAUSE = 1

#Abrindo uma nova aba
pyautogui.hotkey("ctrl", "t")

#Copiando o texto com o pyperclip para evitar erros na cópia do texto
pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing")

#Escrever o endereço a ser acessado
pyautogui.hotkey("ctrl", "v")

#confirmar que esta entrando no site
pyautogui.press("enter")

#definindo tempo de "repouso" do código para dar tempo de carregar a página web
time.sleep(5)
```

```python
#Passo 2: Navegar no sistema e encontrar a pasta base de vendas (no exemplo é a pasta exportar)

# descobrir a posição do mouse para clicar
# pyautogui.position()

#Definir até onde o mouse deve ir para abrir a pasta
pyautogui.click(x=383, y=316, clicks=2)
time.sleep(2)

#selecionar o arquivo
pyautogui.click(x=427, y=402)

#clicando nos pontinhos
pyautogui.click(x=1712, y=198)

#clicando em fazer download
pyautogui.click(x=1565, y=565)

time.sleep(5)
```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos

```python
# Passo 4: Importar a base de vendas pro python
import pandas as pd

tabela = pd.read_excel("Vendas - Dez.xlsx")
display(tabela)
```

```python
# Passo 5: Calcular os indicadores da empresa

faturamento = tabela["Valor Final"].sum()
quantidade = tabela["Quantidade"].sum()
```

### Vamos agora enviar um e-mail pelo gmail

```python
# Passo 6: Emvoar um e-mail para a diretoria com os indicadores de venda

#abrir nova aba
pyautogui.hotkey("ctrl", "t")

# entrar no site do e-mail
pyperclip.copy("mail.google.com")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(5)
#clicar no botão escrever
pyautogui.click(x=86, y=211)

#preencher as informações do e-mail
#pyperclip.copy("SEU E-MAIL AQUI")
#pyautogui.hotkey("ctrl", "v")
pyautogui.write("SEU E-MAIL AQUI")
pyautogui.press("tab") #selecionar o e-mail
pyautogui.press("tab") #mudar para campo de assunto
pyperclip.copy("Relatório de vendas")
pyautogui.hotkey("ctrl", "v")
pyautogui.press("tab") #mudar para o corpo do e-mail

texto = f"""
Prezado, segue o relatório de vendas
Faturamento: R${faturamento:,.2f}
Quantidade: {quantidade:,}
Qualquer dúvida, estou a disposição
Att. Raphael"""

pyperclip.copy(texto)
pyautogui.hotkey("ctrl", "v")

#enviar o e-mail
pyautogui.hotkey("ctrl", "enter")
```
